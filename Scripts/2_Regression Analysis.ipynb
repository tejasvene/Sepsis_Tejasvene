```python
#Age&Sex
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import statsmodels.api as sm
import csv

class glm():

  def __init__(self,path):
    self.df = pd.read_csv(path, sep='\t')
    
  def process(self):
    data = self.df.iloc[:,0:36]
    X= data[['Age','Sex']]
    y= data.iloc[:,7]
    return(X,y)

  def logit_summary (self,X,y):
    logit_model= sm.Logit(y,X)
    result= logit_model.fit()
    RegAnalysis = result.summary2()
    print(RegAnalysis)

    for x in range (0, 2):
      results= result.pvalues[x]
      print(results)

    
  
    return(logit_model,y)

if __name__ == "__main__":
  obj = glm('/content/drive/MyDrive/SergheiLabProjects/Sepsis/Sepsis_Tejas/GeneExpressionAnalysis/CellCompositionData/Cleaned_CellTypePredictionsWithMetadata.tsv')
  #print(data)
  X,y = obj.process()
  logit_model,y = obj.logit_summary(X,y)
  
```
